# Stemming

### Stemming is the process of removing suffixes or prefixes of a word to extract the root word.

New words in English are generally formed by adding a verb, possession or a tense.

Stemming rmeoves these and returns the root word.

But it can be erroneous in cases where removing the prefix or suffix doesnt give the root word.

for example, leaves -> leav , where the root word is leaf
taught -> is considered as a new word even though its the past tense of "teach"




```python
from nltk.stem.porter import PorterStemmer
# Reduce words to their stems
stemmed = [PorterStemmer().stem(w) for w in words]
# printing all stemmed words
print(stemmed)
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
# using snowball stemmer, we can specifiy the language of the text too
stemmer = SnowballStemmer(language = "english")
# Reduce words to their stems
stemmed_by_snowball = [stemmer.stem(word) for word in words]
# printing all stemmed words
print(stemmed_by_snowball)
```

# Comparing porterstemmer and snowball stemmer

```python
comp = []
for i,j in zip(stemmed, stemmed_by_snowball):
    if i != j:
        comp.append(i + " " + j)

# comparing the stemmed words from the diff stemmer libraries and showing the stems that were interpreted differently
print("\n".join(comp))
```
